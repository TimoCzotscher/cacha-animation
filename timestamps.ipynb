```python
import json
import os
import matplotlib.pyplot as plt
```

```python
# load by user json data
data = {}
for filename in os.listdir("data/log-by-user"):
    with open("data/log-by-user/" + filename) as f:
        data[filename.removesuffix(".json")] = json.load(f)
```

```python
data
```

```python
# collect timestamps for each team
timestamps = {}
for team in data.keys():
    timestamps[team] = []
    for entry in data[team]:
        timestamps[team].append(entry['current_location']['timestamp'])
```

```python
# plot timestamps
for team in timestamps.keys():
    plt.scatter(timestamps[team], [team] * len(timestamps[team]))

plt.show()
```
