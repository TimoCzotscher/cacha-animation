```python
import json
import os
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go
```

```python
# load by user json data
data = {}
for filename in os.listdir("data/log-by-user"):
    with open("data/log-by-user/" + filename) as f:
        data[filename.removesuffix(".json")] = json.load(f)
```

```python
teams = data.keys()
```

```python
# collect timestamps for each team
timestamps = {}
for team in teams:
    timestamps[team] = []
    for entry in data[team]:
        timestamps[team].append(entry['current_location']['timestamp'])
```

```python
# plot timestamps using matplotlib
for team in teams:
    plt.scatter(timestamps[team], [team] * len(timestamps[team]), label=team)
plt.show()
```

```python
# plot timestamps using plotly
fig = px.scatter()
for team in teams:
    fig.add_scatter(x=timestamps[team], y=[team] * len(timestamps[team]), mode='markers', name=team)
fig.show()
```
